```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
path = './Accuracies/embedding_shapes_cv_accs.npy'

results = np.load(path)
```

```python
x = 2**np.arange(3, 13)
standard_errors = np.std(results, axis=1)/np.sqrt(results.shape[1])
means = np.mean(results, axis=1)

plt.figure(figsize=(6, 6))
plt.errorbar(x, means, yerr=standard_errors)
plt.title('Embedding Shapes Performance')
plt.xlabel('Embedding Shape')
plt.ylabel('Accuracy')
plt.xscale('log')
plt.savefig('embedding_shape_results.png',dpi=500)
plt.show()
```
