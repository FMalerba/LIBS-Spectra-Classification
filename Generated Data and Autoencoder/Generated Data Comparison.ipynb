```python
import pandas as pd
import numpy as np
from tqdm import tqdm
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
generated_data = np.load('/home/malerba/LIBSqORE/data/NIST_generated/data.npy')
generated_data_labels = np.load('/home/malerba/LIBSqORE/data/NIST_generated/labels.npy')
referenz_df = pd.read_csv('/home/malerba/LIBSqORE/data/190520_mod/chalkosin-4-session_1_01PM-shot(25).csv')

chalkosin_generated_data = generated_data[generated_data_labels == 'chalkosin']

del generated_data, generated_data_labels
```

```python
random_index = np.random.choice(np.arange(0, 12500))
generated_df = chalkosin_generated_data[random_index]
```

```python
plt.figure(figsize=(12, 5))
plt.subplot(1,2,1)
plt.plot(referenz_df['wavelength'], generated_df)
plt.title('Generated 1D Spectrum')
plt.xlabel('Wavelength')
plt.ylabel('Intensity')
plt.subplot(1,2,2)
plt.plot(referenz_df['wavelength'],referenz_df['intensity'])
plt.title('Real 1D Spectrum')
plt.xlabel('Wavelength')
plt.ylabel('Intensity')

plt.savefig('gen_vs_real_data.png',dpi=500)

plt.show()
```
